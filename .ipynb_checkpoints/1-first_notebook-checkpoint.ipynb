```python
def monthly_stats(month, team):
    correct_month = epl_results.loc[f"{month}"]
    games_played = correct_month.loc[(correct_month.HomeTeam==f"{team}") | (correct_month.AwayTeam==f"{team}")].count()[0]
    new_df = correct_month.loc[((correct_month.HomeTeam==f"{team}") & (correct_month.FTR=="H") | (correct_month.AwayTeam==f"{team}") & (correct_month.FTR=="A"))]
    games_won = new_df.FTR.count()
    home_wins = new_df.query('FTR=="H"').FTR.count()
    away_wins = new_df.query('FTR=="A"').FTR.count()
    home_goals_scored = new_df.loc[new_df.HomeTeam==team].FTHG.sum()
    away_goals_scored = new_df.loc[new_df.AwayTeam==team].FTAG.sum()
    my_new_df = pd.DataFrame({
        'month': month,
        'Team': [team],
        'GamesPlayed': [games_played],
        'GamesWon': [games_won],
        'HomeWins': [home_wins],
        'AwayWins': [away_wins],
        'HomeGoalsScored': [home_goals_scored],
        'AwayGoalsScored': [away_goals_scored]
        
        
    })
    my_new_df.set_index('month', inplace=True)
    return my_new_df
#     return new_df.loc[:, ['GamesWon', 'GamesPlayed', 'HomeWins', 'AwayWins']].T
```

# Percentage of goals scored by a big club per big clubs

```python
import math
def monthly_big_teams_stats(month, your_team):
    all_goals = []
    teams = ['Chelsea','Man City', 'Arsenal', 'Man United', 'Tottenham', 'Newcastle', 'Liverpool']
    months = []
    games = []
    gamesWon = []
    homeWins = []
    awayWins = []
    homeGoalsScored = []
    awayGoalsScored = []
    teams2 = []
    games = []
    for team in teams:
        correct_month = epl_results.loc[f"{month}"]
        games_played = correct_month.loc[(correct_month.HomeTeam==f"{team}") | (correct_month.AwayTeam==f"{team}")].count()[0]
        new_df = correct_month.loc[((correct_month.HomeTeam==f"{team}") & (correct_month.FTR=="H") | (correct_month.AwayTeam==f"{team}") & (correct_month.FTR=="A"))]
        games_won = new_df.FTR.count()
        home_wins = new_df.query('FTR=="H"').FTR.count()
        away_wins = new_df.query('FTR=="A"').FTR.count()
        home_goals_scored = new_df.loc[new_df.HomeTeam==team].FTHG.sum()
        away_goals_scored = new_df.loc[new_df.AwayTeam==team].FTAG.sum()
        
        
        months.append(month)
        teams2.append(team)
        games.append(games_played)
        gamesWon.append(games_won)
        homeWins.append(home_wins)
        awayWins.append(away_wins)
        homeGoalsScored.append(home_goals_scored)
        awayGoalsScored.append(away_goals_scored)

    my_new_df = pd.DataFrame({
        'month': months,
        'Team': teams2,
        'GamesPlayed': games,
        'GamesWon': gamesWon,
        'HomeWins': homeWins,
        'AwayWins': awayWins,
        'HomeGoalsScored': homeGoalsScored,
        'AwayGoalsScored': awayGoalsScored
                      })
    my_new_df.set_index('month', inplace=True)
    your_team_goals = my_new_df.loc[my_new_df.Team==your_team].HomeGoalsScored.sum() + my_new_df.loc[my_new_df.Team==your_team].AwayGoalsScored.sum()
    
    
    for team in teams:
        goals = my_new_df.loc[my_new_df.Team==team].HomeGoalsScored.sum() + my_new_df.loc[my_new_df.Team==team].AwayGoalsScored.sum()
        all_goals.append(goals)
    overall = sum(all_goals)
    percentage = (your_team_goals / overall) * 100

    return f"Goals Scored by Big 7 - {overall}.\nGoals Scored by {your_team} - {your_team_goals}.\nOverall Percentage of goals by {your_team} amongst the Big 7 - {math.floor(percentage)}%."
```

```python
print(monthly_big_teams_stats('05/2023', 'Man United'))
```
